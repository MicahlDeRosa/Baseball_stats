```python
#looking at individual players
players = df_3['Player']
players

```

```python
#looking at strike out rate
#SO_Rate = SO / AB
Matt_SO = df_3.loc[10].at['SO']
Matt_AB = df_3.loc[10].at['AB']
Matt_H = df_3.loc[10].at['H']
Matt_HR = df_3.loc[10].at['HR']
Matt_BB = df_3.loc[10].at['BB']
Matt_TB = df_3.loc[10].at['TB']
Matt_SF = df_3.loc[10].at['SF']

SO_Rate1 = Matt_SO/Matt_AB
round(SO_Rate1, 3)
```

```python
#looking at strike out rate
#SO_Rate = SO / AB
Tuffy_SO = df_3.loc[3].at['SO']
Tuffy_AB = df_3.loc[3].at['AB']
Tuffy_H = df_3.loc[3].at['H']
Tuffy_HR = df_3.loc[3].at['HR']
Tuffy_BB = df_3.loc[3].at['BB']
Tuffy_TB = df_3.loc[3].at['TB']
Tuffy_SF = df_3.loc[3].at['SF']


SO_Rate2 = Tuffy_SO/Tuffy_AB
round(SO_Rate2, 3)
```

```python
#looking at BABIP(the batting average on balls hit in play)
#BABIP = (H – HR) / (AB – SO – HR) 

BABIP_matt = (Matt_H - Matt_HR) / (Matt_AB - Matt_SO - Matt_HR + Matt_SF)
round(BABIP_matt, 3)

```

```python
BABIP_tuffy = (Tuffy_H - Tuffy_HR) / (Tuffy_AB - Tuffy_SO - Tuffy_HR + Tuffy_SF)
round(BABIP_tuffy, 3)
```

```python
#homerun rate
HR_rate1 = Matt_HR / (Matt_AB - Matt_SO)
round(HR_rate1, 3)
```

```python
HR_rate2 = Tuffy_HR / (Tuffy_AB - Tuffy_SO)
round(HR_rate2, 3)
```

```python
#isolated Power(ISO)
#(1x2B + 2x3B + 3xHR) / At-bats OR Slugging percentage - Batting average

Matt_SO1 = team.loc[12].at['SO']
Matt_AB1 = team.loc[12].at['AB']
Matt_H1 = team.loc[12].at['H']
Matt_HR1 = team.loc[12].at['HR']
Matt_BB1 = team.loc[12].at['BB']
Matt_TB1 = team.loc[12].at['TB']
Matt_2B1 = team.loc[12].at['2B']
Matt_3B1 = team.loc[12].at['3B']
Matt_AVG1 = team.loc[12].at['AVG']

ISO_matt = (1*Matt_2B1 + 2*Matt_3B1 + 3*Matt_HR1)/ (Matt_AB1-Matt_AVG1)
round(ISO_matt,3)
```
